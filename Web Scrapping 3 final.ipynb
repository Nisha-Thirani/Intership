results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
```

```python
URL=[]
start=0
end=3

for page in range(start,end):
    links=driver.find_elements(By. XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]' )
    for i in links:
        URL.append(i.get_attribute('href'))
nextbutton=driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
nextbutton.click()
print(len(URL))
        
```

```python
Brand_Name=[]
Name=[]
Price=[]
ReEx=[]
Delivery=[]
Availibility=[]

for i in URL:
    driver.get(i)
    time.sleep(2)
    try:
        Brand=driver.find_element(By. XPATH, '//div[@class="a-section a-spacing-small a-spacing-top-small"]/table/tbody/tr[2]/td[2]')
        Brand_Name.append(Brand.text)
    except NoSuchElementException:
        Brand_Name.append('-')
        
    try:
        name=driver.find_element(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
        Name.append(name.text)
    except NoSuchElementException:
        Name.append('-')
        
    try:
        price=driver.find_element(By.XPATH, '//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
    
    try:
        reex=driver.find_element(By.XPATH, '//div[@class="a-column a-span12 a-text-center icon-container scrollable-container celwidget"]/span/div[2]/span')
        ReEx.append(reex.text)
    except NoSuchElementException:
        ReEx.append('-')
        
    try:
        delivery=driver.find_element(By.XPATH, '//div[@class="a-spacing-base"]')
        Delivery.append(delivery.text)
    except NoSuchElementException:
        Delivery.append('-')
        
        
    try:
        availibility=driver.find_element(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
        Availibility.append(availibility.text)
    except NoSuchElementException:
        Availibility.append('-')
        
    
```

```python
print(len(Brand_Name)), len(Name), len(Price), len(ReEx), len(Delivery), len(Availibility)
```

```python
df=pd.DataFrame({'Brand Name':Brand_Name, 'Product Name':Name, "Price":Price, 'Return & Exchange':ReEx, 'Delivery':Delivery, "Availability":Availibility})
df  
```
