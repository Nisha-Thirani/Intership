# 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team andrating.
c) Top 10 ODI bowlers along with the records of their team andrating.

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def odi_teams(url):
    response = requests.get(url)

    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")

    teams = []
    matches = []
    points = []
    ratings = []

    for row in table.tbody.find_all("tr"):
        team = row.find("span", class_="u-hide-phablet").text.strip()
        teams.append(team)

        match = row.find_all("td")[2].text.strip()
        matches.append(match)

        point = row.find_all("td")[3].text.strip()
        points.append(point)

        rating = row.find_all("td")[4].text.strip()
        ratings.append(rating)

    data = {"Team": teams[:10], "Matches": matches[:10], "Points": points[:10], "Rating": ratings[:10]}
    df = pd.DataFrame(data)

    df.index = range(1, len(df)+1)
    return df
```

```python
df = odi_teams("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
df
```

b) Top 10 ODI Batsmen along with the records of their team andrating.

```python
def top_batsmenss(url):
    response = requests.get(url)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    name = []
    team = []
    rate =[]
    a=soup.find('div', 'rankings-block__banner--name').text
    b=soup.find('div','rankings-block__banner--nationality').text[2:5]
    c=soup.find('div','rankings-block__banner--rating').text
    name.append(a)
    team.append(b)
    rate.append(c)
    namesss = []
    for i in soup.find_all('td', 'table-body__cell name'):
        namesss.append(i.text.strip())
    teams = []
    for i in soup.find_all('td', 'table-body__cell nationality-logo'):
        teams.append(i.text.strip())
    rating = []
    for i in soup.find_all('td', 'table-body__cell u-text-right rating'):
        rating.append(i.text.strip())
    names = name + namesss
    teams = team + teams
    rating = rate + rating
    
    df = pd.DataFrame({
        'Batsmen':names,
        'Team':teams,
        'Rating':rating
    })
    df.index = range(1, len(df)+1)
    return df[:10]
```

```python
name = top_batsmenss('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
name
```

c) Top 10 ODI bowlers along with the records of their team andrating.

```python
def get_top_10_ODI_bowlers(url):
    response = requests.get(url)

    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find('table', attrs={'class': 'table'})

    data = []
    rows = table.find_all('tr')
    for row in rows[1:11]:
        cells = row.find_all('td')
        player_name = cells[1].text.strip()
        team = cells[2].text.strip()
        rating = cells[3].text.strip()
        data.append({
            'Player Name': player_name,
            'Team': team,
            'Rating': rating
        })

    df = pd.DataFrame(data)
    df.index = range(1, len(df)+1)
    return df
```

```python
top = get_top_10_ODI_bowlers('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
top
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data framei) Headline
ii) Time
iii) News Link

```python
    file=requests.get("https://www.cnbc.com/world/?region=world")

    soup=BeautifulSoup(file.content, 'html.parser')

    headlines=[]

    for i in soup.find_all('a', class_="LatestNews-headline"):
        headlines.append(i.text)
    
    df=pd.DataFrame({'Headline':headlines})

    span=[]

    for i in soup.find_all('span', class_="LatestNews-wrapper"):
        span.append(i.text)
    
    df1=pd.DataFrame({'Time':span})

    url=[]
 
    for i in soup.find_all('a', class_="LatestNews-headline"):
        url.append(i.get('href'))
    
    df2=pd.DataFrame({'Links':url})
```

```python
result=pd.concat([df,df1,df2], axis=1,join='inner')
result
```

a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
def icc_womens_team(url):
    response = requests.get(url)

    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table")

    teams = []
    matches = []
    points = []
    ratings = []

    for row in table.tbody.find_all("tr"):
        team = row.find("span", class_="u-hide-phablet").text.strip()
        teams.append(team)

        match = row.find_all("td")[2].text.strip()
        matches.append(match)

        point = row.find_all("td")[3].text.strip()
        points.append(point)

        rating = row.find_all("td")[4].text.strip()
        ratings.append(rating)

    data = {"Team": teams[:10], "Matches": matches[:10], "Points": points[:10], "Rating": ratings[:10]}
    df = pd.DataFrame(data)

    df.index = range(1, len(df)+1)

    return df
```

b) Top 10 women’s ODI Batting players along with the records of their team and rating.

```python
def top_10_women_batting_players(url):
   
    response = requests.get(url)

    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find('table', {'class': 'table rankings-table'})

    data = []

    rows = table.find_all('tr')

    for row in rows[1:11]:
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        player = cols[1]
        team = cols[2]
        rating = cols[3]
        data.append([player, team, rating])

    df = pd.DataFrame(data, columns=['Player', 'Team', 'Rating'])

    df.index += 1

    return df
```

```python
wo_batsmen = top_10_women_batting_players('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
wo_batsmen
```

# 7) Write a python program to scrape mentioned details from dineout.co.inand make data framei) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
    page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")

    soup=BeautifulSoup(page.content, 'html.parser')

    name=[]

    for i in soup.find_all('a', class_="restnt-name ellipsis"):
        name.append(i.text)
    
    df=pd.DataFrame({"Restaurant name":name})

    cuisine=[]

    for i in soup.find_all('div', class_="detail-info"):
        cuisine.append(i.text.split('|')[1])
    
    df1=pd.DataFrame({'Cuisine':cuisine})

    loc=[]

    for i in soup.find_all('div', class_="restnt-loc ellipsis"):
        loc.append(i.text)
    
    df2=pd.DataFrame({"Location":loc})

    ratings=[]

    for i in soup.find_all('div', class_="restnt-rating rating-4"):
        ratings.append(i.text)
    
    df3=pd.DataFrame({"Ratings":ratings})

    image_url=[]

    for i in soup.find_all('img', class_="no-img"):
        image_url.append(i.get('data-src'))
    
    df4=pd.DataFrame({"Image URL":image_url})
```

```python
result=pd.concat([df,df1,df2,df3,df4], axis=1,join='inner')
result
```
